```python
class UserAuth:
    def __init__(self, key):
        self.key = key
        self.user_file = "user.txt"

    def _encrypt(self,text):
        return "".join(chr(ord(c)+self.key) for c in text)
    
    def _decrypt(self,text):
        return "".join(chr(ord(c)-self.key) for c in text)
    
    def register(self,username, password):
        encrypted_pwd = self._encrypt(password)

        with open(self.user_file, "a") as file:
            file.write(f"{username}:{encrypted_pwd}\n")

        print("User Registered Successfully")
    
    def delete_user(self, username):
        users = []
        found = False

        try:
            with open(self.user_file, "r") as file:
                users = file.readlines()

            with open(self.user_file, "w") as file:
                for line in users:
                    stored_user, _ = line.strip().split(":")
                    if stored_user != username:
                        file.write(line)
                    else:
                        found = True
            
            if found:
                print("User Deleted Successfully!!")
            else:
                print("UserName not found")

        except FileNotFoundError:
            print("No user File Found")

    def login(self,username,password):
        encrypted_pwd = self._encrypt(password)

        try:
            with open(self.user_file, "r") as file:
                for line in file:
                    stored_user,stored_pwd = line.strip().split(":")

                    if stored_user == username and stored_pwd == encrypted_pwd:
                        return True
        
        except FileNotFoundError:
            print("No User Registered!!")
        
        return False

class FileEncryptor:
    def __init__(self, key):
        self.key = key

    def encrypt_file(self,input_file, output_file):
        stack = []

        with open(input_file, "r") as file:
            for char in file.read():
                stack.append(chr(ord(char)+self.key))
        
        with open(output_file, "w") as file:
            while stack:
                file.write(stack.pop())
        
        print("File Encrypted!!")

    def decrypt_file(self, input_file,output_file):
        queue = []

        with open(input_file, "r") as file:
            for char in reversed(file.read()):
                queue.append(chr(ord(char)-self.key))
            
        with open(output_file,"w") as file:
            for char in queue:
                file.write(char)
        
        print("File Decrypted!!")

class SecureApp:
    def __init__(self):
        self.key = 2
        self.auth = UserAuth(self.key)
        self.encryptor = FileEncryptor(self.key)
        self.logged_in = False

    def menu(self):
        print("\n Secure File Encryption Tool")
        print("1. Register")
        print("2. Login")
        print("3. Encrypt File")
        print("4. Decrypt File")
        print("5. Delete User")
        print("6. Exit")

    def run(self):
        while True:
            self.menu()
            choice = input("Enter Your Choice : ")

            if choice == "1":
                user = input("Username : ")
                pwd = input("Password : ")
                self.auth.register(user,pwd)

            elif choice == "2":
                user = input("Username : ")
                pwd = input("Password : ")
                self.logged_in = self.auth.login(user,pwd)
                print("Login Successful" if self.logged_in else "Invalid Credentials")

            elif choice == "3":
                if self.logged_in:
                    self.encryptor.encrypt_file("input.txt", "encrypted.txt")
                else:
                    print("Access Denied! Login Required")
            
            elif choice == "4":
                if self.logged_in:
                    self.encryptor.decrypt_file("encrypted.txt", "decrypted.txt")

            elif choice == "5":
                if self.logged_in:
                    username = input("Enter Username to Delete : ")
                    self.auth.delete_user(username)
                    self.logged_in = False
                else:
                    print("Access Denied! Login Required")
                    
            elif choice == "6":
                print("Exiting the Program...")
                break 
            else:
                print("Invalid choice")


if __name__ == "__main__":
    app = SecureApp()
    app.run()
```
